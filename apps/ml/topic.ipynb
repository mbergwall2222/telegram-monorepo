```python
import cohere
import numpy as np
from qdrant_client import QdrantClient
from sklearn.metrics.pairwise import cosine_similarity

client = QdrantClient(url="http://qdrant.qdrant.svc:6333")
```

```python
def generate_embedding(description):
    co = cohere.Client("Bl70tSVXSOY9kJqpZs5mRYN4AZHJ7kHao9lOBvF5")
    response = co.embed(texts=[description], model="embed-multilingual-v3.0", input_type="search_document")
    return response.embeddings[0]
```

```python
topic = generate_embedding("""Refund fraud is a deceptive practice where individuals exploit e-commerce return policies to obtain refunds for items they've purchased online. This is typically done by hiring "refunders," individuals who specialize in orchestrating these scams. Common tactics include:

Fraudulent Tracking Number Scans: The scammer provides a fake or unrelated tracking number to the e-commerce company, falsely claiming the item was returned. The tracking number appears legitimate, misleading the company to issue a refund.

Empty Box Scam: The scammer returns the box of the purchased item, but without the actual product inside. This is often done in a way that makes it difficult for the company to detect the fraud until it's too late.

FTID (Fake Tracking ID): Similar to fraudulent tracking number scans, this involves providing a fake or altered tracking ID to prove the return of an item that was never actually sent back.

These tactics cause significant losses to e-commerce businesses and often require sophisticated methods to detect and prevent. They exploit the trust-based system that many online retailers use to handle returns and refunds, taking advantage of the fact that it's often impractical for businesses to thoroughly check every return before issuing a refund.""")
topic
```
